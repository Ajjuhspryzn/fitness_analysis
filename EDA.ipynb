```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Age": np.random.randint(18, 50, 100),
    "Weight": np.random.randint(50, 100, 100),
    "Height": np.random.randint(150, 190, 100),
    "Workout_Duration": np.random.randint(20, 120, 100),
    "Calories_Burned": np.random.randint(150, 800, 100),
    "Heart_Rate": np.random.randint(70, 180, 100)
}

df = pd.DataFrame(data)

df["BMI"] = df["Weight"] / ((df["Height"]/100)**2)

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.scatter(df["Workout_Duration"], df["Calories_Burned"])
plt.title("Workout Duration vs Calories Burned")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.show()
```

```python
sns.histplot(df["Calories_Burned"], kde=True)
plt.title("Calories Distribution")
plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df, x="Workout_Duration", y="Calories_Burned",
                 size="Heart_Rate", color="BMI",
                 title="Interactive Fitness Analysis")
fig.show()
```
